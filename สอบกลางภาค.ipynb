##ข้อ2 หา สส แบบแบ่งเขตและพรรคที่สังกัด ที่ได้คะแนนเลือกตั้งน้อยที่สุด N อันดับ โดยที่ N คือเศษที่ได้จากการหาร รหัสนักศึกษา 2 ตัวสุดท้ายด้วย เลข 5 ยกเว้นคนที่เศษเป็น 0 ให้ N=5 ตัวอย่าง เช่น IDxxxxxx13 N = 3 (30 คะแนน)

```python
def N2(x):
  y = x%5
  if y == 0:
    print(5)
  else:
    print(y)
```
